```python
from datasets import load_dataset
data_files = {'train': 'data/episode_0_episode_19/train-*'}
dataset = load_dataset("Tahahah/PacmanDataset_3", split="train", verification_mode="no_checks")
```

```python
from PIL import Image
import numpy as np
from datasets import load_dataset, Dataset

def create_black_frame(size=(552, 456)):
    return Image.new("RGB", size, "black")



def preprocess_function(batch):
    num_frames = 64
    all_samples = []

    for i in range(len(batch['frame_image'])):
        start_idx = max(0, i - num_frames + 1)
        end_idx = i + 1

        frames = [batch['frame_image'][j] if j >= start_idx else create_black_frame() for j in range(i - num_frames + 1, i + 1)]
        actions = [batch['action'][j] if j >= start_idx else 0 for j in range(i - num_frames + 1, i + 1)]
        next_frames = [batch['next_frame_image'][j] if j >= start_idx else create_black_frame() for j in range(i - num_frames + 1, i + 1)]
        done = batch['done'][i]

        all_samples.append({
            'frames': frames,
            'actions': actions,
            'next_frames': next_frames,
            'done': done
        })

    return all_samples

processed_data = preprocess_function(dataset.select(range(10)))
```

```python
processed_dataset = Dataset.from_dict({
    'frames': [sample['frames'] for sample in processed_data],
    'actions': [sample['actions'] for sample in processed_data],
    'next_frames': [sample['next_frames'] for sample in processed_data],
    'done': [sample['done'] for sample in processed_data]
})
```
